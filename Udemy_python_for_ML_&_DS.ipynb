# Subplots

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

a = np.linspace(0,10,11)
b = a** 4

x = np.arange(0,10)
y = 2 * x

x
```

```python
y
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(x,y)
axes[0][0].plot(a,b)
```

```python
type(axes)
```

```python
axes.shape
```

## SUBPLOTS_ADJUST & suptitle

```python
axes[1][0]
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1)

for ax in axes:
  ax.plot(x,y)

fig.subplots_adjust(wspace = 0.9,hspace =0.4)

axes[0].set_xlabel('X')
axes[0].set_title('first title')

axes[1].set_title('title')
axes[2].set_ylabel('Y')


fig.suptitle('Top Title', fontsize = 24)

fig.savefig('top title', bbox_inches = 'tight')
```

## legend, color, linewidth

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0,11,10)

fig = plt.figure()

fig,axes = plt.subplots(nrows = 1, ncols = 1)

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label = 'one', color = 'red', linewidth = 7, linestyle = '--')
ax.plot(x,x**2, label = 'two' , color = '#b60bd9', lw = 10, ls = ':')

ax.legend(loc = (-1,1))
```

```python
type(ax)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0,11,10)

fig = plt.figure()

fig,axes = plt.subplots(nrows = 1, ncols = 1)

ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x, x + 1, color = 'purple', linewidth = 5)
lines[0].set_dashes([5,2,5,3])
```
